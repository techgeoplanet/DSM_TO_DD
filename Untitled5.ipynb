```python
import pandas as pd

# Read the CSV file containing DMS coordinates
input_file = 'dms_coordinates.csv'
output_file = 'dd_coordinates.csv'

df = pd.read_csv(input_file)

# Function to convert DMS to DD
def dms_to_dd(degrees, minutes, seconds, direction='N'):

    print(degrees,minutes,seconds)
    dd = float(degrees) + float(minutes) / 60 + float(seconds) / 3600
    if direction in ['S', 'W']:
        dd = -dd
    return dd

# Apply the conversion to latitude and longitude columns
df['DD Lat'] = df.apply(lambda row: dms_to_dd(row['Lat_D'], row['Lat_M'], row['Lat_S']), axis=1)
df['DD Lon'] = df.apply(lambda row: dms_to_dd(row['Lon_D'], row['Lon_M'], row['Lon_S']), axis=1)

# Save the result to a new CSV file
df.to_csv(output_file, index=False)

print("Conversion complete. DD coordinates saved to", output_file)
```
